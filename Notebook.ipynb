## Amenity, Crossing, Junction and Traffic Signal - Greatest Correlation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.DataFrame(data[['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)', 'Severity', 'Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal']])
```

```python
#The variables used for the classification model
#Chaning type boolean to integer
df = df.replace([True, False], [1,0])
#Fills the NaN values with average values
df.fillna(df.mean(), inplace=True)
df.info()
```

```python
df_num = pd.DataFrame(df[['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)', 'Severity']])
df_cat = pd.DataFrame(df[['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Severity']])
```

```python
#Training and Test Datasets Split
X = df.drop('Severity', axis=1)
y= df['Severity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#Training and Test Datasets - Numeric
X = df_num.drop('Severity', axis=1)
y= df_num['Severity']

X_train_num, X_test_num, y_train_num, y_test_num = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#Training and Test Datasets - Categorical
X = df_cat.drop('Severity', axis=1)
y= df_cat['Severity']

X_train_cat, X_test_cat, y_train_cat, y_test_cat = train_test_split(X, y, test_size=0.30, random_state=42)
```

## Random Forest Classification

    It makes use of many decision trees that are trained using different subsets of data from our dataset and using random      subset of features in the data
    Since we have a column that has the severity from 1 to 4
    The model is trained such that it classifies the data into 4 classes using the trees
    The final prediction is made by either averaging the predicted probabilities or aggregating the predictions of all trees


```python
#Random Forest Classification - Both Numeric and Categorical Data
clf_base = RandomForestClassifier()

clf_base.fit(X_train, y_train)
y_pred = clf_base.predict(X_test)

print (classification_report(y_test, y_pred, digits = 3))
```

## Classification Report

    Precision: Percentage of correct positive predictions relative to total positive predictions
    Recall: Percentage of correct positive predictions relative to total actual positives
    F1 Score: Weighted harmonic mean of precision and recall (Closer to 1 = Better Model) 
    ----Calculation = [2 * (Precision * Recall) /(Precision + Recall)]
        

```python
#Random Forest Classification - Numeric data only
clf_base_num = RandomForestClassifier()

clf_base_num.fit(X_train_num, y_train_num)
y_pred_num = clf_base_num.predict(X_test_num)

print (classification_report(y_test_num, y_pred_num, digits = 3))
```

```python
#Random Forsest Classification - Categorical data only
clf_base_cat = RandomForestClassifier()

clf_base_cat.fit(X_train_cat, y_train_cat)
y_pred_cat = clf_base_cat.predict(X_test_cat)

print (classification_report(y_test_cat, y_pred_cat, digits = 3))
```

```python
importances = clf_base.feature_importances_

# Get the indices of the features sorted by importance
indices = np.argsort(importances)[::-1]

# Plot the feature importances
plt.figure(figsize=(40, 20))
plt.title("Feature Importance Plot")
plt.bar(range(X_train.shape[1]), importances[indices])
plt.xticks(range(X_train.shape[1]), X_train.columns[indices])
plt.show()
```
